```python
from main import graham_scan, point
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
q = 10**6
w = 10**6
n_values = range(1, 10**6 + 2, 10**4)
exp1_1_times_A = []
exp1_1_times_B = []
for n in n_values:
        # Генерация n точек
    points = []
    for _ in range(n):
        point.x = np.random.uniform(0, q)
        point.y = np.random.uniform(0, w)
        points.append(point(point.x, point.y))
       
        # Замер времени работы алгоритма А
    start = time.time()
    graham_scan(points, "DHeap")
    end = time.time()
    time_A = end - start
    exp1_1_times_A.append(time_A)

        # Замер времени работы алгоритма В
    start = time.time()
    graham_scan(points, "KMerge")
    end = time.time()
    time_B = end - start
    exp1_1_times_B.append(time_B)
    
        # Вывод количества точек в алгоритме
    print("Кол-во отсортированных точек:", n)
```

```python
q_w = 10**6
n_values = range(1, 10**6 + 2, 10**4)
exp1_2_times_A = []
exp1_2_times_B = []

for n in n_values:
        # Генерация n точек
    points = []
    for _ in range(n):
        side = np.random.choice(['top', 'bottom', 'left', 'right'])  
        if side == 'top':
            point.x = np.random.randint(0, q_w)  
            point.y = q_w  
        elif side == 'bottom':
            point.x = np.random.randint(0, q_w)  
            point.y = 0 
        elif side == 'left':
            point.x = 0  
            point.y = np.random.randint(0, q_w)  
        else:
            point.x = q_w  
            point.y = np.random.randint(0, q_w)  
        points.append(point(point.x, point.y))
        
        # Замер времени работы алгоритма А
    start = time.time()
    graham_scan(points, "DHeap")
    end = time.time()
    time_A = end - start
    exp1_2_times_A.append(time_A)

        # Замер времени работы алгоритма В
    start = time.time()
    graham_scan(points, "KMerge")
    end = time.time()
    time_B = end - start
    exp1_2_times_B.append(time_B)

        # Вывод количества точек в алгоритме
    print("Кол-во отсортированных точек:", n)
```
